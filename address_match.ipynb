```python
rl = {}
for row in df_resident_list.itertuples():
    add = str(row._2) + ' ' + row._4.strip() 
    name = row.Name.strip()
    if add in rl.keys():
        rl[add].append(name)
    else:
        rl[add] = [name]
print(len(rl))
print(rl)
```

```python
al = {}
for row in df_assessors.itertuples():
    add = row.Merged.strip()
    owner1 = row._7.strip()
    owner2 = row._8.strip()
    owner3 = row._9.strip()
    if add in al.keys():
        if owner1 and owner1 not in al[add]:
            al[add].append(owner1)
        if owner2 and owner2 not in al[add]:
            al[add].append(owner2)
        if owner3 and owner3 not in al[add]:
            al[add].append(owner3)
    else:
        if owner1:
            al[add] = [owner1]
        if owner2 and owner2 not in al[add]:
            al[add].append(owner2)
        if owner3 and owner3 not in al[add]:
            al[add].append(owner3)
print(len(al))
print(al)
```

```python
mb = {}
for row in df_mailing_bedroom.itertuples():
    add = row._9.strip()
    owner1 = row.Owner.strip()
    owner2 = row._7.strip()
    owner3 = row._8.strip()
    if add in mb.keys():
        if owner1 and owner1 not in mb[add]:
            mb[add].append(owner1)
        if owner2 and owner2 not in mb[add]:
            mb[add].append(owner2)
        if owner3 and owner3 not in mb[add]:
            mb[add].append(owner3)
    else:
        if owner1:
            mb[add] = [owner1]
        if owner2 and owner2 not in mb[add]:
            mb[add].append(owner2)
        if owner3 and owner3 not in mb[add]:
            mb[add].append(owner3)
print(len(mb))
print(mb)
```

```python
def basic_campare(n1, n2):
    for name in n1:
        if name in n2:
            return True
    return False

def name_compare(n1, n2):
    l1 = []
    l2 = []
    for name in n1:
        l1.append(name.split(' '))
    for name in n2:
        l2.append(name.split(' '))
    for arg in l1:
        for arg2 in l2:
            i = 0
            for w in arg:
                if w in arg2:
                    i += 1
            if i >= 2:
                return True
    return False
        
i = 0
add = list(int_add)[5]
print(al[add])
print(rl[add])
print(basic_campare(al[add], rl[add]))
print(name_compare(al[add], rl[add]))
print('\n')
```
