```python
import googlemaps
from datetime import datetime

gmaps = googlemaps.Client(key='AIzaSyBH6Z0lvEOGtgLCrT0rwSqautdBSiLlhBs')

geocoding = pd.DataFrame(columns = ['address', 'owner_occupied', 'longitude', 'latitude'])

addlist = list(address['address'])
oolist = list(address['owner_occupied'])
for i in range(len(addlist)):
    # Geocoding an address
    add = addlist[i]
    geocode_result = gmaps.geocode(add + ', Revere, MA')
    try:
        loc = geocode_result[0]['geometry']['location']
        geocoding = geocoding.append({'address': add, 'owner_occupied': oolist[i], 'longitude': loc['lng'], 'latitude': loc['lat'], }, ignore_index=True)
    except:
        print('Request error.')
geocoding
```
